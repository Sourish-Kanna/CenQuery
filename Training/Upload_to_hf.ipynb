```python
from huggingface_hub import login, create_repo, HfApi

# 1. Login (Will ask for your HF Token)
# Get a WRITE token here: https://huggingface.co/settings/tokens
login()
```

```python
# 2. Configuration
# CHANGE THIS to your desired model name
repo_id = "Sourish-Kanna/CenQuery"
ADAPTER_PATH = "/content/drive/MyDrive/Major_Project/models/CenQuery_Adapter_New_1"
```

```python
# 3. Create Repo & Upload
try:

    print("fk Uploading files...")
    api = HfApi()
    api.upload_folder(
        folder_path=ADAPTER_PATH, # Uploads the files you dragged into Colab
        repo_id=repo_id,
        repo_type="model",
        ignore_patterns=[".config", "sample_data"] # Ignore default Colab junk
    )
    print(f"✅ Success! Adapter is now at: https://huggingface.co/{repo_id}")

except Exception as e:
    print(f"❌ Error: {e}")
```
